```
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style="darkgrid")
```

```
df=sns.load_dataset("iris")
```

```
df.head()
```

```
sns.relplot( x="sepal_length", y="sepal_width",color="r",data=df)
```

```
tips=sns.load_dataset("tips")
```

```
tips.head()
```

```
sns.relplot(x="total_bill", y="tip", data=tips,hue="smoker", style="sex",)
```

```
sns.relplot(x="total_bill", y="tip", data=tips,hue="smoker", style="size",)
```

```
sns.relplot(x="total_bill", y="tip",hue="size", data=tips,palette="ch:r=0.8, l= 0.98")
```

```
sns.relplot(x="total_bill", y="tip",hue="smoker", size="size", data=tips)
```

```
sns.relplot(x="total_bill", y="tip",hue="smoker", size="size", data=tips)

```

```
sns.relplot(x="total_bill", y="tip",hue="smoker", size="size", data=tips,kind="line")
```

```
fmri=sns.load_dataset ("fmri")
```

```
sns.relplot(x="timepoint",y="signal",data=fmri, kind="line", ci=False)
```

```
sns.relplot(x="timepoint",y="signal",data=fmri, kind="line", ci="sd")
```

```
sns.relplot(x="timepoint",y="signal",data=fmri,hue="event",kind="line", ci="sd")
```

```
sns.relplot(x="timepoint",y="signal",data=fmri,hue="region",style="event",kind="line", ci="sd")
```
